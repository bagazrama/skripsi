```python
import pandas as pd

# Memuat data latih dan data uji
train_data = pd.read_csv('C:/skripsi/coding_final/fitur_warna_tomat_HSV.csv')
test_data = pd.read_csv('C:/skripsi/coding_final/uji_fitur_warna_tomat_HSV.csv')

# Distribusi kelas pada data latih
print("Distribusi Kelas pada Data Latih")
print(train_data['Class'].value_counts())

# Distribusi kelas pada data uji
print("\nDistribusi Kelas pada Data Uji")
print(test_data['Class'].value_counts())
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Memuat data fitur warna HSV untuk pelatihan
train_data = pd.read_csv('C:/skripsi/coding_final/fitur_warna_tomat_HSV.csv')

# Memuat data fitur warna HSV untuk pengujian
test_data = pd.read_csv('C:/skripsi/coding_final/uji_fitur_warna_tomat_HSV.csv')

# Gabungkan data pelatihan dan pengujian untuk analisis distribusi
data = pd.concat([train_data, test_data])

# Plot distribusi nilai fitur HSV
features = ['Mean_H', 'Mean_S', 'Mean_V', 'Std_H', 'Std_S', 'Std_V']
for feature in features:
    plt.figure(figsize=(6, 4))
    plt.hist(data[feature], bins=50, color='blue', alpha=0.7)
    plt.title(f'Distribusi Nilai {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frequency')
    plt.show()
```
